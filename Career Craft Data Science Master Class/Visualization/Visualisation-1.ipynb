```python
#analyse the performance of Virat Kohli and Rohit Sharma using line chart
```

```python
#To find the exact name of Rohit Sharma and Virat Kohli
"Rohit" in str(df.Batsman.unique())
sorted(df.Batsman.unique())
```

```python
players = ["Rohit Sharma", "Virat Kohli"]
df_players = df[df['Batsman'].isin(players)]
df_players
```

```python
for player in players:
    player_data = df_players[df_players["Batsman"] == player]
    plt.plot(player_data['match_no'], player_data['Run'], marker = 'o', label = player)

plt.title("Matchwise runs of Viratkohli and Rohit Sharma")
plt.xlabel("Match no")
plt.ylabel("Runs")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#different markers
markers=["o","D"]

i=0

for p in players:
    player_data =df_players[df_players["Batsman"] == p]
    plt.plot(player_data["match_no"] , player_data["Run"] , marker = markers[i], label=player)
    i = i+1
plt.show()
```

```python
#another way
df_kohli = df[df["Batsman"] == "Virat Kohli"]
df_rohit = df[df["Batsman"] == "Rohit Sharma"]

plt.plot(df_kohli["match_no"], df_kohli["Run"], marker="s", linestyle="-", label="Virat Kohli", color="blue")
plt.plot(df_rohit["match_no"], df_rohit["Run"], marker="^", linestyle="-", label="Rohit Sharma", color="red")

plt.title("Match-wise Runs of Virat Kohli & Rohit Sharma")
plt.xlabel("Match Number")
plt.ylabel("Runs Scored")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df
```

```python
#internal homework:
#strike rate (Runs divide by ball for each match) trend over matches >> meaning/insight of the value >> higher strike rate means aggressive inning

df_kohli["Strike Rate"] = (df_kohli["Run"] / df_kohli["Ball"]) * 100
df_rohit["Strike Rate"] = (df_rohit["Run"] / df_rohit["Ball"]) * 100

plt.plot(df_kohli["match_no"], df_kohli["Strike Rate"], marker="o", linestyle="-", label="Virat Kohli", color="blue")
plt.plot(df_rohit["match_no"], df_rohit["Strike Rate"], marker="o", linestyle="-", label="Rohit Sharma", color="red")

plt.title("Match-wise Strike Rate of Virat Kohli & Rohit Sharma")
plt.xlabel("Match Number")
plt.ylabel("Strike Rate")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Boundary count trend

plt.plot(df_kohli["match_no"], df_kohli["4s"], marker="s", linestyle="--", label="Virat Kohli - 4s", color="blue")
plt.plot(df_kohli["match_no"], df_kohli["6s"], marker="^", linestyle="-.", label="Virat Kohli - 6s", color="darkblue")

plt.plot(df_rohit["match_no"], df_rohit["4s"], marker="s", linestyle="--", label="Rohit Sharma - 4s", color="red")
plt.plot(df_rohit["match_no"], df_rohit["6s"], marker="^", linestyle="-.", label="Rohit Sharma - 6s", color="darkred")

plt.title("Boundary Trend of Virat Kohli & Rohit Sharma")
plt.xlabel("Match Number")
plt.ylabel("Count of 4s & 6s")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#task
#make a strike rate column in df, calculate average of each batsman, then plot avg vs average strike rate  
df['strike_rate']=round((df.Run/df.Ball)*100, 2)

batsman_avg = df.groupby("Batsman")["Run"].mean()
batsman_avg
```

```python
batsman_sr_avg = df.groupby("Batsman")["strike_rate"].mean()
batsman_sr_avg
```

```python
plt.scatter(batsman_avg, batsman_sr_avg, color='red', alpha = 0.4) #alpha is for intensity of colour

plt.xlabel("Average Runs")

plt.ylabel("Average Strike Rate")

plt.title("Average Runs vs Average Strike Rate of Batsmen")

plt.grid()

plt.show()
```

```python
#Raise the Bar >> annot the top 5 batsman name here on the plot, so that management can see the name of those batsman on the plot and buy them


#steps:
#create the dataframe 
batsman_stats = pd.DataFrame({"Avg_Runs": batsman_avg, "Avg_SR": batsman_sr_avg})
batsman_stats = batsman_stats.sort_values(by = "Avg_SR", ascending = False)
```

```python
batsman_stats
```

```python
plt.figure(figsize = (18, 10))
plt.scatter(batsman_stats["Avg_Runs"], batsman_stats["Avg_SR"], color = "Red", alpha=0.5)


for i,  player in enumerate(batsman_stats.index[:5]):
    plt.annotate(player, (batsman_stats["Avg_Runs"][i], batsman_stats["Avg_SR"][i]), fontsize=10)
```
