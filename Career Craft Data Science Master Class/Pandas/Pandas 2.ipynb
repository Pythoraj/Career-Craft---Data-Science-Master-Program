```python
#Q. How many female passengers less than 20 years old has paid less than average fare?
titanic[(titanic.Fare < titanic.Fare.mean()) & (titanic.Sex=='female') & (titanic.Age<20)].count()
```

```python
len(titanic[(titanic.Fare > titanic.Fare.mean()) & (titanic.Survived==1) & (titanic.Age<20)])
```
