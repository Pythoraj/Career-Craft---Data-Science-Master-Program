```python
#use case 1 >> sales of stores
#Q.Total sales for each store
data = {
    'Store': ['A', 'B', 'A', 'B', 'A', 'C', 'C'],
    'Sales': [200, 150, 300, 100, 250, 400, 350]
}
df = pd.DataFrame(data)
df
```

```python
df.groupby('Store')['Sales'].sum()
```

```python
#use case 2
data = {
    'Department': ['HR', 'Tech', 'HR', 'Tech', 'Sales', 'Sales'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Salary': [50000, 70000, 52000, 75000, 40000, 45000]
}
df = pd.DataFrame(data)

#with respect each department you want average salary
df.groupby('Department')['Salary'].mean()
```

```python
data = {
    'Class': ['10A', '10A', '10B', '10B', '10C', '10C'],
    'Subject': ['Math', 'English', 'Math', 'English', 'Math', 'English'],
    'Score': [85, 90, 78, 82, 92, 88]
}
df = pd.DataFrame(data)

#with respect to each class calculate average score
df.groupby('Class')['Score'].mean()
```

```python
#with respect to each class calculate average score in each subject
df.groupby(['Class', 'Subject'])['Score'].mean()
```

```python
data = {
    'Customer': ['John', 'Mary', 'Mike', 'Sue', 'Tom'],
    'Age': [25, 40, 25, 30, 40],
    'Purchase_Amount': [200, 300, 150, 100, 250]
}
df = pd.DataFrame(data)
df
#based on different age groups, calculated average purchase amount
```

```python
bins = [20, 30, 40, 50]
labels = ['20-30', '30-40', '40-50']

df['Age_group'] = pd.cut(df['Age'], bins = bins, labels=labels, right=False)
```

```python
df
```

```python
df.groupby('Age_group')['Purchase_Amount'].sum()
```
