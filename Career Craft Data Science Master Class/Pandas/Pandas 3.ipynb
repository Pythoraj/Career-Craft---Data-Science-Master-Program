```python
#merge
#merge needs a common column to be merged while concatenation doesnt need
```

```python
data = {'StoreA':[100, 200, 312, 410, 321],
       'StoreB': [121, 130, 142, 170, 182],
       'StoreC':[111, 121, 123, 124, 191]}
df1 = pd.DataFrame(data)
```

```python
data = {'StoreA':[100, 200, 311, 410, 322],
       'StoreD': [12, 130, 142, 170, 322],
       'StoreE':[111, 124, 123, 124, 124]}
df2 = pd.DataFrame(data)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how = 'inner')
#common column, matching values are used to merge
```

```python
pd.merge(df1, df2, how = 'left')
```

```python
pd.merge(df1, df2, how = 'right')
```

```python
pd.merge(df1, df2, how = 'outer')
#it is call full merge, and it considers the common column all the values from both the datframe
#all the keys from both the dataframe
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how = 'cross')
#cartesian product of the dataframe
#use case of this:
#Analogy, there are four persons in one dataframe and in another dataframe there are places to visit
#if all the person wants to visit all the place, use cross merge
#since we had the same name StoreA in both the column, the merge has changed its name to differentiate
```

```python
#do we need to have common column in both thge dataframe to have cross merge >> No
#store A in 2nd datframe has been changed to StoreF
data = {'StoreA':[100, 200, 312, 410, 321],
       'StoreB': [121, 130, 142, 170, 182],
       'StoreC':[111, 121, 123, 124, 191]}
df1 = pd.DataFrame(data)

data = {'StoreF':[100, 200, 311, 410, 322],
       'StoreD': [12, 130, 142, 170, 322],
       'StoreE':[111, 124, 123, 124, 124]}
df2 = pd.DataFrame(data)


pd.merge(df1, df2, how = 'cross')
```

```python
#what if the common column name is different and we have to do left merge?


data = {'StoreA':[100, 200, 312, 410, 321],
       'StoreB': [121, 130, 142, 170, 182],
       'StoreC':[111, 121, 123, 124, 191]}
df1 = pd.DataFrame(data)

data = {'Store_a':[100, 200, 311, 410, 322],
       'StoreD': [12, 130, 142, 170, 322],
       'StoreE':[111, 124, 123, 124, 124]}
df2 = pd.DataFrame(data)


pd.merge(df1, df2, how = 'left', left_on="StoreA", right_on="Store_a")
#similar thing can be done on other merge if common column name is different
```

```python
#Can we achieve right merge using left in the how parameter of merge
pd.merge(df2, df1, how = 'left', right_on="StoreA", left_on="Store_a")
```

```python
#what if in the merge you have null values

data = {'StoreA':[100, 200, 312, 410, 321],
       'StoreB': [121, 130, 142, 170, 182],
       'StoreC':[111, 121, 123, 124, 191]}
df1 = pd.DataFrame(data)

data = {'StoreA':[100, np.NaN, 311, 410, 322],
       'StoreD': [12, 130, 142, 170, 322],
       'StoreE':[111, 124, 123, 124, 124]}
df2 = pd.DataFrame(data)

pd.merge(df1, df2, how = 'left')

#In this case null value behaves normally as not a common eleement
```

```python
#bucket the age column
#if age< 25, >> young
#if age greater than 25 and less than 45 >> matured
#if age > 45>> old
```

```python
def age_bucket(age):
    if age < 25:
        return 'young'
    elif 25<=age<=45:
        return 'matured'
    else:
        return 'old'
```

```python
df["Age_bucket"] = df["Age"].apply(age_bucket)
```

```python
df
```

```python
#another method
bins = [0, 25, 45, float('inf')] #edges of the bins
labels = ['young', 'matured', 'old']
pd.cut(df["Age"], bins = bins, labels=labels, right=False) #right=False to not to include the right part of bins in the interval
```
