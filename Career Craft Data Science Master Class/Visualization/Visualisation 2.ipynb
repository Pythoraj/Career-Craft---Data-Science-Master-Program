```python
#How much of a batsman total runs came from boundaries? Assume the batsman is Virat Kohli


player = "Virat Kohli"

df_player = df[df["Batsman"] == player]
boundary_runs = (df_player["4s"] * 4 + df_player["6s"] * 6).sum()
other_runs = df_player["Run"].sum() - boundary_runs


plt.figure(figsize=(6,6))
plt.pie([boundary_runs, other_runs], labels=["From Boundaries", "Other Runs"], autopct='%1.1f%%', colors=["pink", "green"])
plt.title(f"Percentage of Runs from Boundaries for {player}")
plt.show()
```

```python
#drawing a horizontal and vertical lines in the plot

plt.figure(figsize=(18,10))
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*50, alpha=0.6, color="purple")

plt.axhline(5, color='red')  # Threshold for tip amount
plt.axhline(10, color='green')  # Higher tip threshold
plt.axvline(20, color='red')  # Threshold for total bill

for i in range(tips.shape[0]):
    plt.text(tips['total_bill'].values[i], tips['tip'].values[i], tips['day'].values[i], fontsize=10)

plt.xlabel("Total Bill Amount ($)")
plt.ylabel("Tip Amount ($)")
plt.title("Total Bill vs. Tip Amount with Table Size Indicated")
plt.show()

# Insights:
# - Red horizontal line at $5 indicates a minimum threshold for tipping behavior.
# - Green horizontal line at $10 represents a higher tip category, possibly linked to generous tipping.
# - Red vertical line at $20 marks the threshold where higher spending may impact tipping habits.
# - Most tips seem to follow an increasing trend with the total bill, though variations exist.
```

```python
#contour plot


import matplotlib.pyplot as plt
import numpy as np

# Define dimensions
Nx, Ny, Nz = 100, 300, 500
X, Y, Z = np.meshgrid(np.arange(Nx), np.arange(Ny), -np.arange(Nz))

# Create fake data
data = (((X+100)**2 + (Y-20)**2 + 2*Z)/1000+1)

kw = {
    'vmin': data.min(),
    'vmax': data.max(),
    'levels': np.linspace(data.min(), data.max(), 10),
}


fig = plt.figure(figsize=(5, 4))
ax = fig.add_subplot(111, projection='3d')


_ = ax.contourf(
    X[:, :, 0], Y[:, :, 0], data[:, :, 0],
    zdir='z', offset=0, **kw
)
_ = ax.contourf(
    X[0, :, :], data[0, :, :], Z[0, :, :],
    zdir='y', offset=0, **kw
)
C = ax.contourf(
    data[:, -1, :], Y[:, -1, :], Z[:, -1, :],
    zdir='x', offset=X.max(), **kw
)

xmin, xmax = X.min(), X.max()
ymin, ymax = Y.min(), Y.max()
zmin, zmax = Z.min(), Z.max()
ax.set(xlim=[xmin, xmax], ylim=[ymin, ymax], zlim=[zmin, zmax])

# Plot edges
edges_kw = dict(color='0.4', linewidth=1, zorder=1e3)
ax.plot([xmax, xmax], [ymin, ymax], 0, **edges_kw)
ax.plot([xmin, xmax], [ymin, ymin], 0, **edges_kw)
ax.plot([xmax, xmax], [ymin, ymin], [zmin, zmax], **edges_kw)

# Set labels and zticks
ax.set(
    xlabel='X [km]',
    ylabel='Y [km]',
    zlabel='Z [m]',
    zticks=[0, -150, -300, -450],
)

# Set zoom and angle view
ax.view_init(40, -30, 0)
ax.set_box_aspect(None, zoom=0.9)

# Colorbar
fig.colorbar(C, ax=ax, fraction=0.02, pad=0.1, label='Name [units]')

# Show Figure
plt.show()
```
